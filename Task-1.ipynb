# Step [1]: Prepare libraries and data

## [1.1] Include important libraries

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression  
from sklearn import metrics  
%matplotlib inline
```

## [1.2] Read data 

```python
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)

s_data.head(5)
```

## [1.3] Preparing the data

```python
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

# Step[2] : Plotting the distribution of Data

```python
s_data.plot(x='Hours', y='Scores', style='*', c='g')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

# Step[3] : Training the Algorithm

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

## - Plotting the regression line

```python
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

# Step[4] : Making Predictions

```python
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
hours = 9.25
hours = np.asarray(hours).reshape((-1 ,1))
own_pred = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

# Step[5] : Evaluating the model 

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```
